# STEP 2: Choosing the right Algorithm

Some things to note:
* Sklearn refferes to the machine learning models, algorithms as estimators.
* Classification problem- Predicting a category(heart Disease or not). Generally represented by clf or model used as a classification estimator
* Regression problem - predicting a number(selling price of a car)

## Picking a machine learning model for a regression model

Getting the built in data set from the Sklearn, we will be using the California Housing Dataset which is a real world dataset

```python
from sklearn.datasets import fetch_california_housing

housing = fetch_california_housing()

housing_df = pd.DataFrame(housing["data"] , columns = housing["feature_names"])
housing_df
```

```python
housing_df["target"] = housing["target"]
housing_df.head()
```

```python
housing_df
```

```python
from sklearn.model_selection import train_test_split
# Import Algorigthm
from sklearn.linear_model import Ridge

# Setup random seed
np.random.seed(42)

# Create the data lable
X = housing_df.drop("target", axis = 1)

y = housing_df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)

model = Ridge()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

Lets try an ensemble model( an ensemble is a combination of smaller models to try and make better prediction than just a simple model


```python
from sklearn.ensemble import RandomForestRegressor

np.random.seed(42)

X = housing_df.drop("target", axis = 1)

y = housing_df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)

model = RandomForestRegressor()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```
