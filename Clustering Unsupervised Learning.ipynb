## Principal Component Analysis (PCA)

 Always Standerize the value before doing the PCA

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

np.random.seed(42)

iris = load_iris()

x, y = iris.data, iris.target
x, y = shuffle(x, y, random_state= 0)
print(f"Features count: {x.shape[1]}")

x[:5, :]

pca = PCA(n_components= 2)

x_reduced = pca.fit_transform(x)
print(f"Features count: {x_reduced.shape[1]}")

x_reduced[:5, :]

plt.scatter(x_reduced[y == 0,0], x_reduced[y==0, 1], c = "r");
plt.scatter(x_reduced[y == 1,0], x_reduced[y==1, 1], c = "b");
plt.scatter(x_reduced[y == 2,0], x_reduced[y==2, 1], c = "g");
```
