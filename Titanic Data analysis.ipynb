## Filling Missing Values

```python
imputer = SimpleImputer(strategy = "most_frequent")
imputer.fit(titanic_df[["Embarked"]])
titanic_df[['Embarked']] = imputer.transform(titanic_df[['Embarked']])
```

```python
titanic_df
```

```python
imputer = SimpleImputer(strategy= "median")
imputer.fit(titanic_df[["Age"]])
titanic_df[['Age']] = imputer.transform(titanic_df[['Age']])
```

```python
titanic_df = titanic_df.drop(["Name", "Ticket", "Fare", "PassengerId", "Cabin"], axis = 1)
```

```python
titanic_df
```

## Converting categorical variables into a numerical format

```python
# OneHotEncoding
nominal_cols = ['Sex', 'Embarked']
encoded_onehot = pd.get_dummies(titanic_df[nominal_cols], dtype=int)
titanic_df.drop(nominal_cols, axis=1, inplace = True)
titanic_df = pd.concat([titanic_df, encoded_onehot], axis=1)
titanic_df.head()
```

```python
target = titanic_df["Survived"]

scaler = StandardScaler()

features = titanic_df.drop('Survived', axis=1) 
normalized_features = scaler.fit_transform(features)

titanic_df = pd.concat([pd.DataFrame(normalized_features, columns=features.columns), target], axis=1)
```

## Labeling the Data_set and Splitting them for Training and Testing


```python
# Selecting the Model
X = titanic_df.drop("Survived", axis = 1)

y = titanic_df["Survived"]

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```
